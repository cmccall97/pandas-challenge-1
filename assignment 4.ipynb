## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
#self explanatory
df['Subtotal'] = df['unit_price'] * df['qty']


df.head(10)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
#Probably a better way to do this, but made a column called temp weight which multiplies weight by qty to check for the 50 pound limit, then applies the lambda func requested, and saves it to Shipping Price.
df['temp_weight'] = df['unit_weight'] * df['qty']
df['Shipping_Price'] = df['temp_weight'].apply(lambda x: x*7 if x>50 else x*10)

df.head()

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
#Thank you chatgpt for explaining how to program sales tax. add subtotal and shipping price, and multiply it by 1*.0925.  = total price
sales_tax = .0925
df['Total_Price'] = (df['Subtotal'] + df['Shipping_Price']) * (1 + sales_tax)
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
#Line cost is a sum of all the costs we've input so far. 

df['Line_cost'] = df['qty'] * df['unit_cost'] + df['Shipping_Price']
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
#self explanatory

df['Profit'] = df['Total_Price'] - df['Line_cost']

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
#filtered_df gets a list of the top 5 clients.  total spending is a variable that: groups by the 5 client ids we got in the previous step, and sums the total price for each of them. resets this output to the index name "total spent"
#and a small formatting adjustment after for readability.

filtered_df = df[df['client_id'].isin(top_five_clients)]
total_spending = filtered_df.groupby('client_id')['Total_Price'].sum().reset_index(name='total_spent')
total_spending['total_spent'] = total_spending['total_spent'].apply(lambda x: '{:.2f}'.format(x))
total_spending
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

#similar to last, if client_id im searching for is in my top_five_clients variable, groups by the client id, and creates 4 variables using the agg function, just summing everything it wants me to. 
#then, created a df, called summary_df, where the 4 previously created variables are given matching column names. not sure resetting the index is helpful here, please let me know in review.

summary = df[df['client_id'].isin(top_five_clients)].groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('Shipping_Price', 'sum'),
    total_revenue=('Total_Price', 'sum'),
    total_profit=('Profit', 'sum')
)

summary_df = summary[[ 'total_units_purchased', 'total_shipping_price', 'total_revenue', 'total_profit']]

summary_df.reset_index()
```

```python
# Format the data and rename the columns to names suitable for presentation.

#money columns is columns related to money. we make a function that converts it to millions using x. used map function to apply to only the money columns found in summary_df. renamed to reflect this.

# Define the money columns. 
money_columns = ['total_shipping_price', 'total_revenue', 'total_profit']


# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    return x / 1000000

# Apply the currency_format_millions function to only the money columns. 
summary_df[money_columns] = summary_df[money_columns].map(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
summary_df = summary_df.rename(columns={
    'total_shipping_price': 'Total Shipping (millions)',
    'total_revenue': 'Total Revenue (millions)',
    'total_profit': 'Total Profit (millions)'
})

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

#self explanatory. sorted and sorted and done. thank you.

sorted_summary_df = summary_df.sort_values(by='Total Profit (millions)', ascending=False)
sorted_summary_df
```
